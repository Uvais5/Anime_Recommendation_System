```python
import pandas as pd 
import numpy as np 
```

```python
anime = pd.read_csv("C:/Machine Learning/z__Project Data Set/anime/animes_2.csv") 
anime
```

```python
# remove duplicate anime 
anime = anime.drop_duplicates() 
```

```python
# This columns is useless for us . so i drop this columns 
anime = anime.drop(columns=["link","members","ranked"])
anime
```

```python
# check word lenght for contain only year 
# this Sep 27, 2011 to Dec 25, 2011 to this 2011
len(anime["aired"][3360])
```

```python
# we only want year of this values 
# so we use sliceing for 2017 
anime["aired"][4][-5:]
```

```python
# now i made a function,
# this function return only year of  all date 
# this date to( Sep 27, 2011 to Dec 25,2011) to this year (2011)
def clean(anime):
    y = []
    for i in anime[-4:]:
        y.append((i))
    return y
```

```python
# Now we apply this function to whole column
anime["aired"] = anime["aired"].apply(clean)
```

```python
# anime["aired"] = anime["aired"].apply(lambda x: [i.replace(",","") for i in x]) 
```

```python
# Then i join spaces words 
anime["aired"] = anime["aired"].apply(lambda x:" ".join(x))
```

```python
# This function remove extra space our raw 
def cc(anime):
    new = ""
    for i in anime:
        if i != " ":
            new += i
    return(new)
# we apply whole column and check the output
anime["aired"] = anime["aired"].apply(cc)
anime["aired"][0]
```

```python
anime[anime["aired"] == "2001"]
```

```python
# now i remove old anime like this .All anime who made in 1999 are no longer in this dataset and orignal dataset
# contain 1945 anime old year anime thats why i have to remove all old anime one by one like this 
an = anime[anime['aired'] == "1999"].index
anime.drop(index=an,inplace=True)
```

```python
# we only want anime thats why we remove this hentai genre anime like this 
an = anime[anime['genre'] == "['Hentai']"].index
anime.drop(index=an,inplace=True)
```

```python
#First we have to convert our column in str 
anime["score"] = anime["score"].astype(str)
```

```python
# Then i create a function.This function return only one word rating like this 3 , 4
# i use for loop to use in all raws and anime[0] is gave first values in raw.
# if i have this values 3.55 and i use anime[0] output is 3 
def clean1(anime):
    y = ""
    for i in anime[0]:
        y=(i)
    return y
```

```python
# We apply this fucntion in whole Column like this
anime["score"] = anime["score"].apply(clean1)
```

```python
# Now i have one word ratings. 
# So i remove 1 to 5 ratings anime like this 
a = anime[anime['score'] == "4"].index
anime.drop(index=a,inplace=True)
```
