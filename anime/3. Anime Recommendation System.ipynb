```python
# NLTK Library gave losts of functions 
# When i use porterstemmer words like love and loving  is convert lov 
# This function is best for better results
import nltk
from nltk.stem.porter import PorterStemmer
# CAll the function 
ps = PorterStemmer() 
# made a function for convert whole column 
def stem(a):
    # empty list for append values
    y = [] 
    # for loop for split every values and work every words one by one 
    for i in a.split():
        # append the converted words for each time
        y.append(ps.stem(i))
        # (Join function)after split i have to join the words
    return " ".join(y)
```
